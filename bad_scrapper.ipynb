```python
import time
import requests
import json
from bs4 import BeautifulSoup as bs
from tqdm import tqdm_notebook as tn
from lxml import html
import pandas as pd
```

```python
web_link='https://masahub.com/' #funny website but don't waste time looking around it.... :) 
hub_data=requests.get(web_link,verify=True) # don't run it and if you really want to set verify to False
print(hub_data.status_code)
```

```python
soup=bs(hub_data.text,features='html.parser')
#print(soup.prettify())
total_pages=int(soup.find('div',{'class':'pager'}).find_all('a',{'class':'page-numbers'})[-2].string)
```

```python
duration=[]
title=[]
page_link=[]
download_link=[]
image_link=[]

print('Careful: each page will take about 60 seconds to process, Give a number between 1 and',total_pages,'.') # got 5s after checking some tests
how_many_pages=input('How many pages do you want to process? ')
print('ETA:',60*int(how_many_pages),'seconds.')

time1=time.time()

page_counter=0
while(page_counter<int(how_many_pages)):
    time3=time.time()
    print(page_counter+1,')   Page No. --> ',str(page_counter+1),' <-- is being processed ...') #to keep track of loops
    for link in soup.find('div',{'class':'main l170'}).find_all('li',{'class':'thumi'}):
        page_link.append(link.find('a',{'class':'title'}).get('href'))
        duration_string=link.find('div',{'class':'duration'}).string.strip().split(':')
        duration.append(float(duration_string[0]+'.'+duration_string[1]))
        foo1=link.img
        image_link.append(foo1.get('src'))
        title.append(foo1.get('alt'))
        time5=time.time()
        dwn_link=page_link[-1]
        data=requests.get(dwn_link,verify=False)
        foo1=bs(data.text,'html.parser')
        foo2=foo1.find_all('div',{'class':'downLink'})[-1].find('a',{'class':'btn'}).get('href')
        download_link.append(foo2)
        time6=time.time()
        link_number=len(download_link) # to indicate which link is being downloaded
        print('Download link no.(',link_number,') fetched in ðŸ‘‰',round(time6-time5,1),'seconds.')
    time4=time.time()
    page_counter+=1
    print('Completed in ðŸ‘‰',round(time4-time3,1),'seconds.')# as the all are taking time in seconds ig
    web_link='https://masahub.com/page/'+str(page_counter+1)+'/'
    soup=bs(requests.get(web_link,verify=False).text,'html.parser')
    
time2=time.time()
print('ðŸ‘‰ It took us',round(time2-time1,1),'seconds, to process your request of processing', how_many_pages,'pages from IMDB webpage. ðŸ‘ˆ')
```

```python
data_df=pd.DataFrame({
    'Title':title,
    'Duration':duration,
    'Img Link':image_link,
    'Download Link':download_link,
    'Page Link':page_link
})
data_df.sort_values('Duration',inplace=True,axis=0)
data_df.reset_index(inplace=True,drop=True)
```
